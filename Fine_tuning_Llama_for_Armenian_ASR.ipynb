 **Why Llama?**

*Llama* is a highly efficient model by Meta, which has been extensively used for various natural language processing (NLP) tasks, including speech-to-text (ASR).

*Llama 2*, in particular, offers strong performance on a range of benchmarks and is available for open use, making it a good choice for fine-tuning on your custom dataset. Llama's architecture is designed to work efficiently, especially for resource-constrained environments, which aligns well with your goal of training an ASR model on Armenian speech data.

```python
# Use a pipeline as a high-level helper

from transformers import pipeline

#Llama 2 7B
generator = pipeline(
    "text-generation",
    model="meta-llama/Llama-2-7b-hf",
    device=0  #use 0 if using a GPU; set device=-1 for CPU
)

prompt = "Who are you?"
output = generator(prompt, max_length=50, do_sample=True)
print(output[0]["generated_text"])
```
