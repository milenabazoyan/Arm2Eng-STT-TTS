```python
import torchaudio
from nemo.collections.asr.models import EncDecHybridRNNTCTCBPEModel
from IPython.display import Audio, display
```

```python
#Load the model from Hugging Face
asr_model = EncDecHybridRNNTCTCBPEModel.from_pretrained(
    model_name="milenabazoyan/fastconformer-hybrid-arm-asr"
)
```

```python
#Load sample audio
audio_path = "/content/common_voice_hy-AM_41864771.wav"
waveform, sample_rate = torchaudio.load(audio_path)
```

```python
#Resample if not 16kHz
if sample_rate != 16000:
    from torchaudio.transforms import Resample
    waveform = Resample(orig_freq=sample_rate, new_freq=16000)(waveform)
```

```python
display(Audio(audio_path))
```

```python
transcription = asr_model.transcribe([waveform.squeeze().numpy()])
print("Armenian Transcription:", transcription[0].text)
```
